## MobileNet

```python
import keras
import numpy as np
import itertools
import matplotlib.pyplot as plt
%matplotlib inline

from keras import backend as K
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras.models import Model
from keras.applications import imagenet_utils
from sklearn.metrics import confusion_matrix
```

```python
mobile = keras.applications.mobilenet.MobileNet()
```

### Initial Implementation Testing

```python
def prepare_image(file):
    img_path = "data/"
    img = image.load_img(img_path+file,target_size=(224,224))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

```python
from IPython.display import Image
Image(filename='data/1.png', width=300)
```

```python
preprocessed_image = prepare_image('1.png')
predictions = mobile.predict(preprocessed_image)
results = imagenet_utils.decode_predictions(predictions)
results
```

#### Importing the Datasets

```python
train_path = 'data/train/'
valid_path = 'data/valid/'
test_path = 'data/test/'
```

```python
## PARAMETERS
tsize = (224,224) #target size
bsize = 10 #batch size
class_list = ['dogs','cats']
```

```python
train_batches = ImageDataGenerator(preprocessing_function=keras.applications.mobilenet.preprocess_input).flow_from_directory(
    train_path, target_size=tsize, batch_size=bsize, classes=class_list)

valid_batches = ImageDataGenerator(preprocessing_function=keras.applications.mobilenet.preprocess_input).flow_from_directory(
    valid_path, target_size=tsize, batch_size=bsize, classes=class_list)

test_batches = ImageDataGenerator(preprocessing_function=keras.applications.mobilenet.preprocess_input).flow_from_directory(
    test_path, target_size=tsize, batch_size=bsize, shuffle=False, classes=class_list)
```

#### Selecting the Layers

```python
mobile.summary()
```

```python
## PARAMETERS
num_last_layers = 6 #specify this to include from the 1st layer up to last ? layer
num_output_nodes = 2 
```

```python
x = mobile.layers[-num_last_layers].output
predictions = Dense(num_output_nodes, activation='softmax')(x)
model = Model(inputs=mobile.input, outputs=predictions)
```

```python
model.summary()
```

```python
layers_tobe_trained = 5

for layer in model.layers[:-layers_tobe_trained]:
    layer.trainable = False
```

#### Training

```python
## PARAMETERS
lrate = 0.0001
epochstep_tra = 4
epochstep_val = 2
```

```python
#Compilation
#Optimizer=Adam
model.compile(Adam(lr=lrate), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#Fitting
model.fit_generator(train_batches, steps_per_epoch=epochstep_tra,
                   validation_data = valid_batches, validation_steps=epochstep_val, epochs=30, verbose=2)
```
